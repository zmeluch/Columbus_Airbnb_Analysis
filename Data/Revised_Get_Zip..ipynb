```python
# Import Dependencies

import geopy
import pandas as pd
```

```python
# Make Listings Data from AirBnB a DataFrame

listings_df = pd.read_csv("Revised_Columbus_Listings.csv", index_col=0)
```

```python
# Show DataFrame

listings_df
```

```python
# Start Geolocator from GeoPY

geolocator = geopy.Nominatim(user_agent="kar")
```

```python
# Function to Retrieve Zip from Latitude and Longitude Coordinates in Listings DataFrame

def get_zip_code(x):
    location = geolocator.reverse("{}, {}".format(x['latitude'],x['longitude']))
    return location.raw['address']['postcode']

listings_df['zipcode'] = listings_df.apply(lambda x: get_zip_code(x), axis = 1)
```

```python
# Show Listings DataFrame with Zip 

listings_df
```

```python
# Export Dataframe as .csv File 

listings_df.to_csv('Revised_Columbus_Listings_with_Zip.csv')
```

```python
# Create dependencies for SQL Alchemy

from sqlalchemy import create_engine
from config import db_password
```

```python
# Add the information to send the data to PostgresSQL
    
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5433/Air_BnB"
    
engine = create_engine(db_string)
    
listings_df.to_sql(name="listings", con=engine)
```
