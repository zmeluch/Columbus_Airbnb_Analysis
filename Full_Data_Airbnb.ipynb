```python
def change_room(r_type):
    if r_type == "Entire home/apt":
        return 1
    else:
        return 0
airbnb_df["room_type"] = airbnb_df["room_type"].apply(change_room)
airbnb_df.head(20)
```

```python
model = KMeans(n_clusters=3, random_state=5)
model
```

```python
airbnb_df.drop(columns=["neighbourhood"], inplace=True)
```

```python
airbnb_df.drop(columns=["calculated_home_listings_count"], inplace=True)
```
